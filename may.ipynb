```python
import pandas as pd
import numpy as np
from sklearn import linear_model
df=pd.read_excel('export.xlsx',index_col=0,nrows=480)
df
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn import linear_model
satellite_list=df['satellite_ws10'].tolist()
gs_list=df['gs_ws10'].tolist()
xs=np.array(satellite_list,dtype=np.float64)
ys=np.array(gs_list,dtype=np.float64)

satellite_gs=linear_model.LinearRegression()
satellite_gs.fit(xs.reshape(-1,1),ys)
regression_line=satellite_gs.predict(xs.reshape(-1,1))
style.use('seaborn')
plt.scatter(xs,ys,alpha=0.6,color='green',s=75)
plt.plot(xs,regression_line,label='Best Fit Line',color='orange',linewidth=4)
plt.xlabel('Satellite wind speed (m/sec)')
plt.ylabel('Ground station wind speed (m/sec)')
plt.show()
```
